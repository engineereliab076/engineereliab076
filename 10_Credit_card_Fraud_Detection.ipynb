Import the deopendencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#load the csv data to pd dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first 5 rows
credit_card_data.head()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking no of missing values
credit_card_data.isnull().sum()
```

```python
#distribution of legit and fraudulent transaction
credit_card_data['Class'].value_counts()
```

```python
#This datset was highly unbalanced.....0, normal transaction and 1 ...,fraudulent transaction
```

```python
#seaparating data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape, fraud.shape)
```

```python
#getting statistical measures of data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values of both transactions
credit_card_data.groupby('Class').mean()
```

Dealing with unbalanced data.
Under- sampling

Build a sample data containing similar no of values for both legut and fraud transaction thats...

492...for fraud

```python
legit_sample = legit.sample(n=492)
```

```python
#concanate 2 df
new_dataset = pd.concat([legit_sample,fraud]
            , axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

split data

```python
#split data into features and target

x = new_dataset.drop(columns='Class', axis=1)
y = new_dataset['Class']
```

```python
print(x, y)
```

Train test split

```python
x_train, x_test, y_train, y_test =  train_test_split(x,y, test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

Model Evaluation

```python
#accuracy score for train data
x_train_pred = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_pred,y_train)
```

```python
print(training_data_accuracy)
```

```python
#accuracy score for test data
x_test_pred = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_pred,y_test)
```
